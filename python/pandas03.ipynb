## 77. 如何在 Pandas 中使用 `plot` 函数画出柱状图？

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    'name':['Tom','Jerry','Mickey','Minnie','Donald'],
    'age':[25,23,28,27,26],
    'score':[80,90,75,85,95]
})
df
```

```python
df.plot(kind='bar',x='name',y='score')
plt.show
```

## 78. 如何在 Pandas 中使用 `plot` 函数画出折线图？

```python
df = pd.read_csv('../../log/loglizer/data/HDFS/anomaly_label.csv',encoding='utf-8')
df
```

```python
import numpy as np
X = np.arange(100)
X
```

```python
y = np.random.randint(10,100,100)
plt.plot(X,y)
```

```python
from matplotlib.font_manager import FontManager
fm = FontManager()
mat_fonts = set(f.name for f in fm.ttflist)
for i in mat_fonts:
    print(i)
print(len(mat_fonts))
```

```python
# 设置字体 显示汉字
plt.rcParams['font.sans-serif'] = 'Hiragino Sans GB'
# 取消使用 unicode 的负号
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.title('这是一个测试')
plt.xlabel('X')
plt.plot(X,y)
plt.ylabel('y')
```

## 80. 如何在Pandas中使用plot函数画出箱线图

```python
data = {
    'A':np.random.normal(0,1,100),
    'B':np.random.normal(2,1,100),
    'C':np.random.normal(3,2,100)
}
df = pd.DataFrame(data=data)
df
```

```python
df.plot.box()
```
